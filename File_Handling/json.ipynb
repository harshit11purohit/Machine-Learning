```python
#  READING & WRITING JSON
import json

data = {
    "name": "Harshit",
    "role": "Developer",
    "skills": ["Python", "File Handling"]
}

with open("data.json", "w") as file:
    json.dump(data, file, indent=4) # indent=4 makes it look pretty!

with open("data.json","r") as file:
    loaded_data = json.load(file)
    print(f"Loaded User: {loaded_data['name']}")
```

```python

import json

my_dict = {"name": "Harshit", "role": "Dev"}

# --- FILES (dump / load) 
with open("data.json", "w") as f:
    json.dump(my_dict, f, indent=4)

# 2. load(): Read FILE into a dictionary
with open("data.json", "r") as f:
    data_from_file = json.load(f)


# --- STRINGS (dumps / loads) - Notice the 's'! ---

# 3. dumps(): Convert dictionary to a STRING
json_string = json.dumps(my_dict, indent=4)
print(json_string)

# 4. loads(): Convert STRING back to a dictionary
data_from_string = json.loads(json_string)
print(data_from_string)
print("Dictionary successfully converted to string and back!")
```

```python
# ==========================================
# 2. DOWNLOADING JSON FROM AN API
# ==========================================

import requests

url = "https://jsonplaceholder.typicode.com/users/1"

response = requests.get(url)  # returns status code
print(response)
# Check if the request was successful (Status Code 200)
if response.status_code == 200:
    user_data = response.json() 
    
    print(f"Name: {user_data['name']}")
    print(f"City: {user_data['address']['city']}")
else:
    print("Failed to get data.")
```

```python
# ==========================================
# 3. SAFELY LOADING JSON (ERROR HANDLING)
# ==========================================


import json
import os

# Let's create a deliberately broken JSON file
with open("broken.json", "w") as f:
    f.write("{ 'name': 'Harshit', }") # Invalid JSON (single quotes, trailing comma)

file_path = "broken.json"

if os.path.exists(file_path):
    with open(file_path, "r") as f:
        try:
            data = json.load(f)
            print("Successfully loaded data.")
        except json.JSONDecodeError as e:
            print("❌ ERROR: The JSON file is corrupted or formatted incorrectly!")
            print(f"Details: {e}")
            
            # Here, you might choose to assign default data so your program doesn't die
            data = {} 
            print("Fallback to empty dictionary.")
```

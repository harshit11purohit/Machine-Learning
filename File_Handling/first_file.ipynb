```python
with open("files.txt",'w') as f:
    f.write("hello guys\n")
    
    
    
with open("files.txt",'r') as r:
    content=r.readlines()
    print(content)
```

```python
lines = [
    "Line 1: Hello, Harshit!\n",
    "Line 2: Welcome to Python.\n",
    "Line 3: File handling is easy.\n"
]

with open("files.txt", "a") as file:
    for line in lines:
        file.write(line)

with open("files.txt",'r') as r:
    content=r.readlines()
    print(content)
```

```python
with open("files.txt", "a") as file:
    file.writelines(lines)
```

```python
with open("files.txt", "r") as file:
    content = file.read()
    print("--- Entire File ---")
    print(content)
```

```python
new_lines = ["Apple: hp\n", "Banana\n", "Cherry\n"]

with open("files.txt", "a") as file:
    file.write("--- My Grocery List ---\n")
    
    file.writelines(new_lines)
    
print("Successfully created and wrote to files.txt!")
```

```python

# 1. READ AND WRITE ("r+")
# Note: The file MUST already exist for "r+" to work.

with open("files.txt", "r+") as file:
    content = file.read()
    print("Old Content:", content)
    
    # 2. Move the cursor back to the top
    file.seek(0) 
    
    # 3. Write new data at cursor point
    file.write("UPDATED SETTINGS\n")
```

```python

# 4. READING CSV DATA
import csv

# Writing to a CSV
with open("employees.csv", "w", newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Department", "Salary"]) # Header
    writer.writerow(["Harshit", "Engineering", "85000"])

# Reading from a CSV
with open("employees.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row) # Prints each row as a Python list
```

```python

# 5. FILE MANAGEMENT (RENAME & DELETE)
import os

with open("temp.txt", "w") as f:
    f.write("I am temporary.")

# 1. Renaming a file
if os.path.exists("temp.txt"):
    os.rename("temp.txt", "permanent.txt")
    print("File renamed!")

# 2. Deleting a file (Always check if it exists first!)
if os.path.exists("permanent.txt"):
    os.remove("permanent.txt")
    print("File safely deleted!")
```

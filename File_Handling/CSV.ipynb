```python
# 4. READING CSV DATA
import csv

# Writing to a CSV
with open("employees.csv", "w", newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Department", "Salary"]) # Header
    writer.writerow(["Harshit", "Engineering", "85000"])

# Reading from a CSV
with open("employees.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row) # Prints each row as a Python list
```

```python

# ==========================================
# 1. WRITING CSV FILES (Using DictWriter)
# ==========================================
import csv

# dictionary
employees = [
    {"Name": "Harshit", "Role": "Developer", "Salary": 85000},
    {"Name": "Alice", "Role": "Data Scientist", "Salary": 95000},
    {"Name": "Bob", "Role": "Manager", "Salary": 105000}
]

# We must define the column headers exactly as they appear in the dictionary keys
fields = ["Name", "Role", "Salary"]

# 'newline=""' prevents blank rows from appearing between data on Windows
with open("company_data.csv", "w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=fields)
    
    writer.writeheader() # Writes the top row (Name, Role, Salary)
    writer.writerows(employees) # Writes all the data at once

print("CSV successfully created!")
```

```python
import pandas as pd

# 1. Load the entire CSV into a powerful "DataFrame" (like a virtual Excel sheet)
df = pd.read_csv("company_data.csv")

print("--- Raw Data ---")
print(df)
print("\n")

high_earners = df[df["Salary"] > 90000]

print("--- High Earners Only ---")
print(high_earners)
print("\n")

# 3. Give everyone a 10% raise using vectorized math (instant calculation)
df["Salary"] = df["Salary"] * 1.10

# 4. Save the updated data back to a NEW CSV file (index=False hides row numbers)
df.to_csv("company_data_updated.csv", index=False)
print("Updated CSV saved with new salaries!")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(n_samples=1000,n_features=2,centers=3,random_state=1)
```

```python
# y not used in unsupervised learning
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=.3)
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
km.fit(x_train)
```

```python
y_labels=km.predict(x_test)
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(x_train,km.labels_)
```

```python
km.inertia_ 
#wcss distances
```

```python
wcss=[]
for k in range(1,10):
     km=KMeans(n_clusters=k)
     km.fit(x_train)
     wcss.append(km.inertia_)
     
wcss
```

```python
# plot the elbow curve
```

```python
plt.plot(range(1,10),wcss)
plt.ylabel("students")
```

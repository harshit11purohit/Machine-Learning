```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
print(load_iris)
```

```python
ld=load_iris()
ld
```

```python
ld.data
```

```python
ld.feature_names
```

```python
df=pd.DataFrame(ld.data,columns=ld.feature_names)
df
```

```python
y=df['target']=ld.target
df
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
y=df.target
x=df.drop('target',axis=1)
```

```python
x
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,train_size=.7)
```

```python
x_test.shape,x_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy')
classifier
```

```python
classifier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.Figure(figsize=(12,30))
tree.plot_tree(classifier,filled=True)
```

```python
y_pred=classifier.predict(x_test)
from sklearn.metrics import accuracy_score,classification_report,r2_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
# Post-pruning removes branches that add little predictive power.
#Post-pruning is a technique used to reduce overfitting by cutting back a fully grown decision tree after training.
#observe descion tree plot  last split is overfitiing we divide to get exactly pure ones 
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier2=DecisionTreeClassifier(criterion='entropy',max_depth=2)
classifier2.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.Figure(figsize=(12,30))
tree.plot_tree(classifier2,filled=True)
```

```python
y_pred2=classifier2.predict(x_test)
from sklearn.metrics import accuracy_score,classification_report,r2_score
accuracy_score(y_test,y_pred2),
print(classification_report(y_test,y_pred2))
```

```python
# pre runing using hyper paramter tuning
```

```python
parameter= { 'criterion':['gini','entropy','log_loss'],
            'splitter':['best','random'],
            'max_depth':[1,2,3,4,5,6,7,10],
            'max_features':['sqrt','log2','auto']
           
           }
```

```python
from sklearn.model_selection import GridSearchCV
clf=DecisionTreeClassifier()
model=GridSearchCV(clf,param_grid=parameter,cv=5,scoring="accuracy")
model
```

```python
model.fit(x_train,y_train)
```

```python
model.best_estimator_
model.best_score_
model.best_params_
```

```python
y_pred3=model.best_estimator_.predict(x_test)
accuracy_score(y_test,y_pred3)
```

```python
from sklearn.model_selection import RandomizedSearchCV
clf=DecisionTreeClassifier()
model3=RandomizedSearchCV(clf,param_distributions=parameter,cv=5,scoring="accuracy")
model3
model3.fit(x_train,y_train)
y_pred4=model3.best_estimator_.predict(x_test)
accuracy_score(y_test,y_pred4)

```

```python
import pandas as pd 
import numpy as np
no_class_0=900
no_class_1=100

class_0={'features': np.random.normal(0,1,900),
         'feature2':np.random.normal(0,1,900),
         'target':[0]*no_class_0
         }
df1= pd.DataFrame(class_0)
df1
```

```python
class_1={'features': np.random.normal(3,1,100),
         'feature2':np.random.normal(3,1,100),
         'target':[1]*no_class_1
         }
df2= pd.DataFrame(class_1)
df2
```

```python
df=pd.concat([df1,df2]).reset_index(drop=True)
df.target.value_counts()
df
```

```python
df_majority=df[df.target==0]
df_minority=df[df.target==1]
```

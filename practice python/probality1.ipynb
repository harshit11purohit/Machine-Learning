```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create sample data
data = {
    'Math': np.random.randint(50, 100, 20),
    'Physics': np.random.randint(45, 95, 20),
    'Chemistry': np.random.randint(40, 90, 20),
    'Biology': np.random.randint(55, 100, 20)
}

df = pd.DataFrame(data)

# Calculate correlation matrix
corr_matrix = df.corr()

print("Correlation Matrix:\n", corr_matrix)

# Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import scipy.stats as stats

# Sample survey data
data = {
    'Gender': ['Male', 'Male', 'Male', 'Female', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Preference': ['Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No']
}

df = pd.DataFrame(data)

# Create a contingency table using crosstab
contingency_table = pd.crosstab(df['Gender'], df['Preference'])
print("Contingency Table:\n", contingency_table)

# Perform Chi-Square Test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print("\nChi-Square Statistic:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)

# Decision at alpha = 0.05
alpha = 0.05
if p < alpha:
    print("\nReject H0: There is a significant association between Gender and Preference")
else:
    print("\nFail to reject H0: No significant association between Gender and Preference")
```

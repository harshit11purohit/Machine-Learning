```python
# Custom exception class for salary-related errors
class exception_salary(Exception):
    def __init__(self, message):
        self.message = message
        super().__init__(message)


# Function to validate salary
def validate_salary(salary):
    if salary <= 0:
        # Raise custom exception when salary is invalid
        raise exception_salary("zero salary or very low")
    
    # Use 'and' instead of '&' (bitwise AND)
    elif salary > 40000 and salary < 100000:
        return "average salary"
    
    else:
        return "high salary"


try:
    # Taking user input
    salary = int(input("Enter salary: "))
    
    # Calling the function
    result = validate_salary(salary)
    print(result)

# ERROR FIX:
# except must catch the EXCEPTION CLASS, not the function name
except exception_salary as e:
    print("the error is as", e)

# Handles non-integer input
except ValueError as e:
    print("the error is there", e)
```

```python
import logging

# Configure logging
logging.basicConfig(
    filename="mylogging.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

listy = [1, 'helo', 22, [2, 'harshit'], 11, 9, [9, 'deepak']]
l1 = []
l2 = []

logging.info("Starting processing of list")

for i in listy:
    logging.info(f"Processing element: {i} (type: {type(i).__name__})")

    if type(i) == list:
        logging.info("Element is a list. Processing sub-elements.")
        for j in i:
            logging.info(f"Sub-element: {j} (type: {type(j).__name__})")

            if type(j) == int:
                l1.append(j)
                logging.info(f"Integer added to l1: {j}")
            else:
                l2.append(j)
                logging.info(f"Non-integer added to l2: {j}")

    elif type(i) == str:
        l2.append(i)
        logging.info(f"String added to l2: {i}")

    else:
        l1.append(i)
        logging.info(f"Integer added to l1: {i}")

logging.info("Processing completed")
logging.info(f"Final integer list (l1): {l1}")
logging.info(f"Final string list (l2): {l2}")
```

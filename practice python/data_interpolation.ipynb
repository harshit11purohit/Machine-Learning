```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.array([1,3,5,7,9,11])
y=np.array([2,4,6,8,10,12])

df=pd.DataFrame([x,y])
df

```

```python
x_new=np.linspace(1,5,50)
df2=pd.DataFrame([x_new])
df2
```

```python
y_interpolation=np.interp(x_new,x,y)
plt.scatter(x_new,y_interpolation)
plt.show
```

```python
import pandas as pd

df = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [2, None, 6, None, 10]
})

df
```

```python
df['y_interp'] = df['y'].interpolate(method='linear')
df
```

```python
plt.scatter(df['x'], df['y_interp'])
plt.show
```

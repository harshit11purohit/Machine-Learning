```python
df=sns.load_dataset("tips")
df
```

```python
import numpy as np
import pandas as pd

x = np.mean(df['total_bill'])   # mean
y = np.std(df['total_bill'])    # standard deviation

listy = list(df['total_bill'])

newx = []
for i in listy:
    z = (i - x) / y
    newx.append(z)

df1 = pd.DataFrame(newx, columns=['total_bill_zscore'])
df1
```

```python
x = np.mean(df['total_bill'])   # mean
y = np.std(df['total_bill'])    # standard deviation

zdata = (df['total_bill'] - x) / y
standraded=pd.DataFrame(zdata)
standraded
```

```python
from sklearn.preprocessing import MinMaxScaler
cols = ['total_bill', 'tip']

scaler = MinMaxScaler()
df[cols] = scaler.fit_transform(df[cols])

df.head()
```

```python
df['total_bill_norm'] = (
    df['total_bill'] - df['total_bill'].min()
) / (
    df['total_bill'].max() - df['total_bill'].min()
)
df.total_bill_norm
```

```python
import logging
import os

print("Current working directory:")
print(os.getcwd())

logging.basicConfig(
    filename="mylogging_warning.log",
    level=logging.WARNING,   # Higher than INFO
    format="%(asctime)s - %(levelname)s - %(message)s",
    force=True               # Required in Jupyter
)

logging.warning("Starting list processing at WARNING level")

listy = [1, 'helo', 22, [2, 'harshit'], 11, 9, [9, 'deepak']]
l1 = []
l2 = []

for i in listy:
    logging.warning(f"Processing element: {i} (type: {type(i).__name__})")

    if isinstance(i, list):
        logging.warning("Element is a list. Processing sub-elements.")
        for j in i:
            if isinstance(j, int):
                l1.append(j)
                logging.warning(f"Integer added to l1: {j}")
            else:
                l2.append(j)
                logging.warning(f"String added to l2: {j}")

    elif isinstance(i, str):
        l2.append(i)
        logging.warning(f"String added to l2: {i}")

    else:
        l1.append(i)
        logging.warning(f"Integer added to l1: {i}")

logging.warning("Processing completed")
logging.warning(f"Final integer list (l1): {l1}")
logging.warning(f"Final string list (l2): {l2}")

print("Execution finished")
```

```python
import threading
import time
from datetime import datetime

# Function each thread runs
def run_thread(name, delay):
    # Wait before starting
    time.sleep(delay)
    print(f"{datetime.now().strftime('%H:%M:%S')} - Thread {name} started after {delay} sec")
    
    # Simulate work
    time.sleep(2)
    print(f"{datetime.now().strftime('%H:%M:%S')} - Thread {name} finished")

# Threads and their start delays
threads_info = [
    ("A", 1),
    ("B", 2),
    ("C", 3)
]

threads = []

# Create and start threads
for name, delay in threads_info:
    thread = threading.Thread(target=run_thread, args=(name, delay))
    threads.append(thread)
    thread.start()

# Wait for all threads to finish
for thread in threads:
    thread.join()

print(f"{datetime.now().strftime('%H:%M:%S')} - All threads finished")
```

```python
import os
import time
import threading
import urllib.request

# List of URLs
url_list = [
    "https://www.w3.org/TR/PNG/iso_8859-1.txt",
    "https://www.learningcontainer.com/wp-content/uploads/2020/04/sample-text-file.txt",
    "https://www.w3.org/TR/PNG/example.html"
]

# Create download folder
os.makedirs("downloads", exist_ok=True)

# Function to download file
def download_file(url):
    filename = url.split("/")[-1]
    file_path = os.path.join("downloads", filename)

    print(f"Starting download: {filename}")
    urllib.request.urlretrieve(url, file_path)
    print(f"Finished download: {filename}")

threads = []

start_time = time.time()

# Create and start threads
for url in url_list:
    t = threading.Thread(target=download_file, args=(url,))
    threads.append(t)
    t.start()

# Wait for all threads
for t in threads:
    t.join()

end_time = time.time()

print(f"All downloads completed in {end_time - start_time:.2f} seconds")
```

```python
import multiprocessing
import time

# Sample data
numbers = [2, 6, 8, 9, 11]

# Function 1: compute square
def square(x):
    time.sleep(0.5)  # simulate heavy computation
    result = x * x
    print(f"Square of {x} is {result}")
    return result

# Function 2: compute double
def double(x):
    time.sleep(0.5)
    result = x * 2
    print(f"Double of {x} is {result}")
    return result

# Function to run both functions in parallel
def process_numbers(nums):
    with multiprocessing.Pool(processes=2) as pool:
        # Map square function
        squares = pool.map(square, nums)
        # Map double function
        doubles = pool.map(double, nums)
    return squares, doubles

# Run multiprocessing
if __name__ == "__main__":
    squares_result, doubles_result = process_numbers(numbers)

    print("\nSquares List:", squares_result)
    print("Doubles List:", doubles_result)
```

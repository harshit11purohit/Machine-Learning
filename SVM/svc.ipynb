```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
x,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
pd.DataFrame(x)
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=1)
```

```python
from sklearn.svm import SVC
clf=SVC(kernel='linear')
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(y_test,y_pred))
accuracy_score(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
params ={'C':[0.1,0.2,0.3,1,2,3,50,100],
         'gamma':[1,.1,.2,.01,.001,.003],
         'kernel':['linear']}
    
```

```python
grid=GridSearchCV(SVC(),param_grid=params,cv=5,verbose=3)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred_grid=grid.best_estimator_.predict(x_test)
y_pred_grid
```

```python
print(classification_report(y_test,y_pred_grid))
accuracy_score(y_test,y_pred_grid)
```

```python
y_pred_grid2=grid.predict(x_test)
print(classification_report(y_test,y_pred_grid2))
accuracy_score(y_test,y_pred_grid2)
```

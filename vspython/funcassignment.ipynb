```python
def reversestring(string):
    n = len(string)
    y = n // 2
    string = list(string)   # strings are immutable, so convert to list
    
    for i in range(y):
        string[i], string[n - i - 1] = string[n - i - 1], string[i]
    
    return ''.join(string)

str1 = "harshit"
result = reversestring(str1)
print(result)
```

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True


# Print primes from 1 to 200
for n in range(1, 201):
    if is_prime(n):
        print(n, end=" ")
```

```python
primes = []
for num in range(2, 201):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        primes.append(num)

print(primes)
```

```python
def read(files):
    for line in files:
        a = "".join(line.split())   # remove spaces
        yield a

files = ["helo guys", "this is awesome", "good men boys", "i am the best", "i have hgh testosterone"]

for i in read(files):
    print(i, end=" ")
```

```python
def read(files):
    for line in files:             # outer loop → each sentence
        a = ""                     # store concatenated version
        for ch in line:            # inner loop → each character
            if ch != " ":          # skip spaces
                a += ch
        yield a                    # yield result for each line


files = ["helo guys", "this is awesome", "good men boys", "i am the best", "i have hgh testosterone"]

for i in read(files):
    print(i, end=" ")
```

```python
def remove(strings):
    chars = [ch for ch in strings]   # make list manually
    for ch in chars[:]:
        if ch in ['a', 'e', 'i', 'o', 'u']:
            chars.remove(ch)
    ans = ""
    for c in chars:
        ans += c
    return ans

strin = "harshitpurohit"
result = remove(strin)
print(result)
```

```python
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    result = "".join(filter(lambda ch: ch not in vowels, s))
    return result

string = "Harshit Purohit"
print(remove_vowels(string))
```

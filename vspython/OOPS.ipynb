```python
class Avengers:
    def __init__(self, strong_avengers, new_avengers, almighty):
        self.strong_avengers = strong_avengers
        self.new_avengers = new_avengers
        self.almighty = almighty

    def extract_strong(self):
        print("Strong Avengers:", self.strong_avengers)

    def extract_new(self):
        print("New Avengers:", self.new_avengers)

    def extract_almighty(self):
        print("Almighty Avengers:", self.almighty)


# --- Example Data ---
strong = ["Iron Man", "Thor", "Hulk", "Captain America"]
new = ["Spider-Man", "Doctor Strange", "Black Panther"]
almighty = ["Captain Marvel", "Scarlet Witch"]

# --- Create an object ---
team = Avengers(strong, new, almighty)

# --- Extract Data ---
team.extract_strong()
team.extract_new()
team.extract_almighty()
```

```python
from abc import ABC, abstractmethod

class pwskills(ABC):  # Abstract class
    @abstractmethod
    def studentdetails(self):
        pass

    @abstractmethod
    def studentmarks(self):
        pass

    @abstractmethod
    def studentattendance(self):
        pass


class datascience(pwskills):
    def studentdetails(self):
        print("Data Science Students")

    def studentmarks(self):
        print("Marks: 850/1000")

    def studentattendance(self):
        print("Attendance: 90%")


class dataanalyst(pwskills):
    def studentdetails(self):
        print("Data Analyst Students")

    def studentmarks(self):
        print("Marks: 900/1000")

    def studentattendance(self):
        print("Attendance: 95%")


# Creating objects
ds = datascience()
da = dataanalyst()

# Calling methods
print("=== Data Science ===")
ds.studentdetails()
ds.studentmarks()
ds.studentattendance()

print("\n=== Data Analyst ===")
da.studentdetails()
da.studentmarks()
da.studentattendance()
```

```python
class cars:
    def __init__(self, raftar, taquat):
        self.__speed = raftar
        self.__power = taquat

    def setspeed(self, raftar):
        if raftar < 0:
            self.__speed = 0   # Cannot set negative speed
        else:
            self.__speed = raftar

    def getpower(self):
        return self.__power


c1 = cars(1000, 999)

c1.setspeed(1200)
print(c1.getpower())
```

```python
class Cars:
    def __init__(self, raftar, taquat):
        self.__speed = raftar     # Private attribute
        self.__power = taquat     # Private attribute

    def setspeed(self, raftar):
        if raftar < 0:
            self.__speed = 0      # Negative speed not allowed
        else:
            self.__speed = raftar

    def getspeed(self):
        return self.__speed       # Getter for speed

    def getpower(self):
        return self.__power       # Getter for power


# Creating object
c1 = Cars(1000, 999)

# Updating and accessing values
c1.setspeed(1200)
print("Speed:", c1.getspeed())
print("Power:", c1.getpower())
```

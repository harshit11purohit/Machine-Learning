```python
import keyword

x = input("Enter a word: ")

if x in keyword.kwlist:
    print("True → It is a Python keyword.")
else:
    print("False → It is not a Python keyword.")
```

```python
def check_keyword(x, keywords):
    size = len(keywords)
    for i in range(size):
        if x == keywords[i]:
            return True
    return False
import keyword

x = input("Enter a word: ")
if check_keyword(x, keyword.kwlist):
    print(f"'{x}' is a Python keyword.")
else:
    print(f"'{x}' is not a Python keyword.")
```

```python
# Take input from user
x = input("Enter a value: ")

# Convert to integer
try:
    y = int(x)
except ValueError:
    y = "Cannot convert to integer"

# Convert to float
try:
    z = float(x)
except ValueError:
    z = "Cannot convert to float"

# Convert to boolean
a = bool(x)  # empty string -> False, non-empty string -> True

# Display results
print("\nConverted Values:")
print("Integer:", y)
print("Float  :", z)
print("Boolean:", a)
```

```python
# Original list with string numbers
x = ["1", "2", "3", "4"]

print("Original list:", x)
print("Original types:", [type(i) for i in x])

# Convert string elements to integers
y = [int(i) for i in x]
print( y)
print([type(i) for i in y])

# Convert integer elements to float
z = [float(i) for i in y]
print("\nAfter converting to floats:", z)
print("Types after float casting:", [type(i) for i in z])

# Convert integer elements to string
a = [str(i) for i in y]
print("\nAfter converting back to strings:", a)
print("Types after string casting:", [type(i) for i in a])

```

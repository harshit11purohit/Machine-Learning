```python
class Bank:
    
    # THIS METHOD RUNS WHEN OBJECT IS CREATED STORES MONEY
    def __init__(self, amount):
        self.amount = amount

    # deposit money
    def deposit(self, money):
        self.amount = money+self.amount
        print(" balance: ", self.amount)
        print(" deposited ",money)

    # withdraw money
    def withdraw(self, money):
        if money <= self.amount:
            self.amount = self.amount- money
            print("Withdrawn:", money)
        else:
            print("Insufficient balance")

    # check balance
    def check_balance(self):
        print("Current Balance:", self.amount)


# object created will be used to access method of a class
b = Bank(5000)

# operations
b.deposit(2000)
b.withdraw(1500)
b.check_balance()
```

```python
class Book:
    def __init__(self, author, name):
        self.author = author
        self.name = name

    def show_book(self):
        print(f"Book: {self.name}, Author: {self.author}")


class Sales(Book):
    def __init__(self, author, name, copies):
        super().__init__(author, name)
        self.copies=copies ## We pass the parameters required by the parent class constructor.        self.copies = copies

    def show_sales(self):
        print(f"Copies sold: {self.copies}")


b1 = Sales("JK Rowling", "Harry Potter", 100)

b1.show_book()  
b1.show_sales() 
```

```python

# Method overriding happens when a child class provides
# its own implementation of a method that already exists
# in the parent class (same method name).

# --------------------------------
# Example 1: Basic Method Overriding
# --------------------------------

class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):          # overriding parent method
        print("Dog barks")

d = Dog()
d.speak()     # Child method runs instead of parent

#. Used for runtime polymorphism.
```

```python
# Example 2: Using super() with overriding
# --------------------------------

class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        super().speak()      # call parent method
        print("Dog barks")

d = Dog()
d.speak()
```

```python
class Animal:
    def __init__(self, name):
        self.name = name
        print("Parent constructor called")

    def speak(self):
        print("Animal makes sound")


class Dog(Animal):

    def __init__(self, name, breed):
        super().__init__(name)   # calling parent __init__..we pass only what needed for parent
        self.breed = breed
        print("Child constructor called")

    def speak(self):
        print(self.breed, "barks")


d = Dog("Tommy", "Labrador")
d.speak()
```

```python
# hierarchial INHERITANCE 

# Hierarchical inheritance means multiple child classes inherit from one parent class.
```

```python

class Parent:
    def show(self):
        print("This is Parent class")

class Child1(Parent):
    def display1(self):
        print("This is Child1")

class Child2(Parent):
    def display2(self):
        print("This is Child2")

c1 = Child1()
c2 = Child2()

c1.show()      # inherited from Parent
c1.display1()

c2.show()      # inherited from Parent
c2.display2()
```

```python
# ==============================
# HYBRID INHERITANCE
# ==============================

# Combination of more than one type of inheritance
# (Multiple + Hierarchical together)

class A:
    def methodA(self):
        print("Class A")

class B(A):          # Hierarchical part
    def methodB(self):
        print("Class B")

class C(A):          # Hierarchical part
    def methodC(self):
        print("Class C")

class D(B, C):       
    def methodD(self):
        print("Class D")

d = D()
d.methodA()
d.methodB()
d.methodC()
d.methodD()
```

```python
# ==============================
# ABSTRACTION (OOP CONCEPT)
# ==============================

# An abstract class = a special type of parent (base) class.
#    It is created at the top level, and other classes inherit from it.
# Abstraction means hiding implementation details
# and showing only essential features to the user.
# using ABC (Abstract Base Class)
#  abstraction usually involves inheritance and method overriding

from abc import ABC, abstractmethod

class Vehicle(ABC):        # Abstract class
    
    @abstractmethod
    def start(self):       # abstract method (no body)
        pass


class Car(Vehicle):        # Child class
    def start(self):
        print("Car starts with key")

class Bike(Vehicle):
    def start(self):
        print("Bike starts with button")


c = Car()
c.start()

b = Bike()
b.start()
```

```python
# Abstraction → hides internal implementation and shows only functionality.
# Abstract class cannot be instantiated directly.
# Child class must implement all abstract methods.
# Implemented using abc module (ABC, abstractmethod).
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass
    
    def fuel_type(self):
        print("Vehicle uses fuel or electricity")


class Car(Vehicle):

    def start(self):  
        print("Car starts with key")


class Bike(Vehicle):

    def start(self):
        print("Bike starts with button")


c = Car()
c.start()
c.fuel_type()

b = Bike()
b.start()
b.fuel_type()
```

```python
from abc import ABC, abstractmethod


class Payment(ABC):

    # Common logic (NOT hidden)
    def payment_success(self, amount):
        print(f"Payment of ₹{amount} successful")

    # Abstract method (implementation hidden)
    @abstractmethod
    def pay(self, amount):
        pass


class CreditCard(Payment):

    def pay(self, amount):
        print("Processing credit card payment...")
        self.payment_success(amount)


# -------- CHILD CLASS 2
class UPI(Payment):

    def pay(self, amount):
        print("Processing UPI payment...")
        self.payment_success(amount)


# -------- USING THE CLASSES --------
c = CreditCard()
c.pay(500)

print("------------------")

u = UPI()
u.pay(1000)
```

```python
##   POLYMORPHISM

# The same function or method behaves differently depending on the object or data.
# In Python, polymorphism is mainly achieved using:
# 1. Method Overriding (same method, different implementation in child class)
# 2. Operator Overloading (same operator works differently for different data)
# 3. Function Polymorphism (same function works with different object types)
```

```python
# method overriding
class Animal:
    def speak(self):
        print("Animal sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")

a = Dog()
B=Cat()
a.speak()
B.speak()
 #Dog barks, Cat meows (same method name, different behavior)
```

```python
# method overloading
class Calculator:
    
    def add(self, a, b=0, c=0):
        return a + b + c

c = Calculator()

print(c.add(5))        # 5
print(c.add(5, 10))    # 15
print(c.add(5, 10, 20))# 35
```

```python
#  The __balance variable is private (note the double underscore __).
#  This means it cannot be accessed directly from outside the class
#  They are used when you need controlled access to class data, not just direct variable access
```

```python
class BankAccount:
    
    def __init__(self, balance):
        self.__balance = balance  # private variable

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(self.__balance)
        else:
            print("Invalid deposit amount!")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(self.__balance)
        else:
            print("Insufficient funds or invalid amount.")

account = BankAccount(1000)
print(account.get_balance())  # Access via getter
account.deposit(500)
print(account.get_balance())
account.withdraw(2000)         # Will show error, can't withdraw more than balance
```

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance  # public variable

    # deposit method
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(self.balance)
        else:
            print("Invalid deposit amount!")

    # withdraw method
    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(self.balance)
        else:
            print("Insufficient funds or invalid amount.")

account = BankAccount(1000)
print(account.balance)  # Direct access, no getter needed
account.deposit(500)
print(account.balance)
account.withdraw(2000)   # Will show error
```

```python
# %% [markdown]
# ### What is a Class Method?
# A class method is a method that belongs to the class itself, rather than to any particular instance (object).
# 
# * It can access or modify **class-level variables**, not instance-level variables.
# * Defined using the `@classmethod` decorator.
# * The first parameter is usually called `cls` (short for "class"), instead of `self`.

# works with class variables ..  no object made ..  x=class_name...method--->x...variable 
# can also be called by instance 
            # acc1 = BankAccount(1000)
            # acc.show_bank_name
```

```python
class MyClass:
    class_variable = 5 # belongs to the class

    @classmethod
    def show_class_variable(cls):
        print(f"Class variable value is : {cls.class_variable}")
        
MyClass.show_class_variable()
```

```python
class BankAccount:
    bank_name = "ABC Bank"  # class variable

    def __init__(self, balance):
        self.balance = balance  # instance variable

    @classmethod
    def show_bank_name(cls):
        print(f"Bank name is {cls.bank_name}")


BankAccount.show_bank_name()  # Access via class


account = BankAccount(1000)
account.show_bank_name()      # Can also access via instance/object
```

```python
class BankAccount:
    bank_name = "ABC Bank"  # class variable

    def __init__(self, balance):
        self.balance = balance  # instance variable

    # Class method to show the bank name
    @classmethod
    def show_bank_name(cls):
        print(f"Bank name is: {cls.bank_name}")

    # Class method as an alternative constructor
    @classmethod
    def create_zero_balance_account(cls):
        return cls(0)  # creates a new BankAccount object with balance 0

BankAccount.show_bank_name() 


acc1 = BankAccount(1000)
print(f"Account 1 balance: {acc1.balance}") 

# Creating an account using class method (alternative constructor)
acc2 = BankAccount.create_zero_balance_account()
print(f"Account 2 balance: {acc2.balance}")  # Output: 0

# Class method can also be called via an instance
acc1.show_bank_name()
```

```python

class Product:
    discount_rate = 0.0  
    
    def __init__(self, name, price):
        self.name = name
        self.price = price
        
    def get_final_price(self):
        return self.price * (1 - self.discount_rate)
        
    @classmethod
    def apply_storewide_discount(cls, new_rate):
        # 3. THE CLASS METHOD
        # Here, we use 'cls' to modify the shared class-level variable.
        # We do NOT use 'self', because this affects the whole blueprint.
        cls.discount_rate = new_rate
        print(f"--> Storewide discount updated to {new_rate * 100}%!")


item1 = Product("Laptop", 1000)
item2 = Product("Headphones", 100)

print("--- BEFORE DISCOUNT ---")
print(f"{item1.name} price: ${item1.get_final_price()}")
print(f"{item2.name} price: ${item2.get_final_price()}")

# Call the class method directly on the Class itself
print("\n--- CALLING CLASS METHOD ---")
Product.apply_storewide_discount(0.20) # Apply a 20% discount

# Look at how modifying the class variable affected ALL existing instances
print("\n--- AFTER DISCOUNT ---")
print(f"{item1.name} price: ${item1.get_final_price()}")
print(f"{item2.name} price: ${item2.get_final_price()}")
```

```python
# ### What is a Static Method?
# A static method is a method that belongs to a class's namespace, but it does not have access to the class itself or its instances.
# 
# * It **cannot** access or modify class-level variables (`cls`) or instance-level variables (`self`).
# * Defined using the `@staticmethod` decorator.
# * It does **not** take a mandatory first parameter (no `self` or `cls` is passed automatically).
# * It behaves exactly like a regular independent function, but is placed inside a class because it logically makes sense for it to live there.
```

```python
class MyClass:
    @staticmethod
    def greet(name):
        print(f"Hello, {name}!")

# Call the static method directly on the Class, passing the argument..no cls no self no instance needed
MyClass.greet('harshit') 
```

```python
class Employee:
    # Class-level variable
    company_name = "TechCorp"
    
    def __init__(self, name):
        # Instance-level variable
        self.name = name

    # Instance/ object method (uses 'self')
    def get_details(self):
        return f"Employee: {self.name} at {self.company_name}"

    # CLASS METHOD (uses 'cls')
    @classmethod
    def change_company_name(cls, new_name):
        cls.company_name = new_name
        print(f"Company name officially changed to {new_name}")

    # 3. STATIC METHOD (no 'self' or 'cls')
    @staticmethod
    def is_workday(day_of_week):
    
        if day_of_week in ["Saturday", "Sunday"]:
            return False
        return True

emp1 = Employee("Harshit")

# Calling Instance Method
print(emp1.get_details()) 

# Calling Static Method
print(f"Is Monday a workday? {Employee.is_workday('Monday')}")

# Calling Class Method
Employee.change_company_name("GlobalTech")
print(emp1.get_details())
```

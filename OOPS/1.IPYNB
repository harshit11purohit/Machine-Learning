```python
class Bank:
    
    # THIS METHOD RUNS WHEN OBJECT IS CREATED STORES MONEY
    def __init__(self, amount):
        self.amount = amount

    # deposit money
    def deposit(self, money):
        self.amount = money+self.amount
        print(" balance: ", self.amount)
        print(" deposited ",money)

    # withdraw money
    def withdraw(self, money):
        if money <= self.amount:
            self.amount = self.amount- money
            print("Withdrawn:", money)
        else:
            print("Insufficient balance")

    # check balance
    def check_balance(self):
        print("Current Balance:", self.amount)


# object created will be used to access method of a class
b = Bank(5000)

# operations
b.deposit(2000)
b.withdraw(1500)
b.check_balance()
```

```python
class Books:
    
    def __init__(self, book_list):
        self.book_list = book_list

    def extract(self):
        print("Book Details:\n")
        for author, name, copies in self.book_list:
            print(f"Book: {name} | Author: {author}")

    def extract2(self):
        print("\nCopies Sold:\n")
        for author, name, copies in self.book_list:
            print(f"{name} → {copies} copies")

data = [
    ('jk', 'harry potter', 100),
    ('hp', 'lord of rings', 50),
    ('np', 'games of thrones', 500)
]

b1 = Books(data)

b1.extract()
b1.extract2()
```

```python
class Books:
    
    def __init__(self, author, name, copies):
        self.author = author
        self.name = name
        self.copies = copies
    
    def extract(self):
        print(f"The name of book is {self.name} and author is {self.author}")
        
    def extract2(self):
        print(f"The number of copies sold is {self.copies}")


# creating objects (jk, hp, np)
b1 = Books('jk', 'harry potter', 100)
b2 = Books('hp', 'lord of rings', 50)
b3 = Books('np', 'games of thrones', 500)

# calling methods
b1.extract()
b1.extract2()

b2.extract()
b2.extract2()

b3.extract()
b3.extract2()
```

```python
# INHERITANCE


# Super() is used to call the parent class methods or constructor.“Go to parent class and run its constructor.”
# It is mainly used for constructors (__init__) or overriding methods
# ** We pass the parameters required by the parent class constructor.
```

```python
''' Child object created
        ↓
Child __init__ runs
        ↓
super().__init__() called
        ↓
Parent __init__ runs 
        -> THEN child method runs '''
```

```python
class Animal:
    def speak(self):
        print("Animal sound")

class Dog(Animal):
    def bark(self):
        print("Dog bark")

d = Dog()
d.bark()
d.speak()
```

```python
# basic inheritance

class Animal:
    def __init__(self):
        print("Animal constructor")

class Dog(Animal):
    def __init__(self):
        super().__init__()
        print("Dog constructor")

d = Dog() 
```

```python
class Book:
    def __init__(self, author, name):
        self.author = author
        self.name = name

    def show_book(self):
        print(f"Book: {self.name}, Author: {self.author}")


class Sales(Book):
    def __init__(self, author, name, copies):
        super().__init__(author, name)  ## We pass the parameters required by the parent class constructor.        self.copies = copies

    def show_sales(self):
        print(f"Copies sold: {self.copies}")


b1 = Sales("JK Rowling", "Harry Potter", 100)

b1.show_book()  
b1.show_sales()  
# child class runs first
```

```python

# Method overriding happens when a child class provides
# its own implementation of a method that already exists
# in the parent class (same method name).

# --------------------------------
# Example 1: Basic Method Overriding
# --------------------------------

class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):          # overriding parent method
        print("Dog barks")

d = Dog()
d.speak()     # Child method runs instead of parent

#. Used for runtime polymorphism.
```

```python
# Example 2: Using super() with overriding
# --------------------------------

class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        super().speak()      # call parent method
        print("Dog barks")

d = Dog()
d.speak()
```

```python
class Animal:
    def __init__(self, name):
        self.name = name
        print("Parent constructor called")

    def speak(self):
        print("Animal makes sound")


class Dog(Animal):

    def __init__(self, name, breed):
        super().__init__(name)   # calling parent __init__..we pass only what needed for parent
        self.breed = breed
        print("Child constructor called")

    def speak(self):
        print(self.breed, "barks")


d = Dog("Tommy", "Labrador")
d.speak()
```

```python
## multiple inheritance

#  Multiple inheritance means a child class inherits from more than one parent class.
```

```python
class Father:
    def skill1(self):
        print("Driving skill")

class Mother:
    def skill2(self):
        print("Cooking skill")

class Child(Father, Mother):
    def skill3(self):
        print("Coding skill")

c = Child()

c.skill1()   # from Father
c.skill2()   # from Mother
c.skill3()   # own method
```

```python
class A:
    def __init__(self):
        print("Class A constructor")

class B:
    def __init__(self):
        print("Class B constructor")

class C(A, B):
    def __init__(self):
        super().__init__()
        super().__init__()
        print("Class C constructor")

c = C()
```

```python
# hierarchial INHERITANCE 

# Hierarchical inheritance means multiple child classes inherit from one parent class.
```

```python

class Parent:
    def show(self):
        print("This is Parent class")

class Child1(Parent):
    def display1(self):
        print("This is Child1")

class Child2(Parent):
    def display2(self):
        print("This is Child2")

c1 = Child1()
c2 = Child2()

c1.show()      # inherited from Parent
c1.display1()

c2.show()      # inherited from Parent
c2.display2()
```

```python
# ==============================
# HYBRID INHERITANCE
# ==============================

# Combination of more than one type of inheritance
# (Multiple + Hierarchical together)

class A:
    def methodA(self):
        print("Class A")

class B(A):          # Hierarchical part
    def methodB(self):
        print("Class B")

class C(A):          # Hierarchical part
    def methodC(self):
        print("Class C")

class D(B, C):       
    def methodD(self):
        print("Class D")

d = D()
d.methodA()
d.methodB()
d.methodC()
d.methodD()
```

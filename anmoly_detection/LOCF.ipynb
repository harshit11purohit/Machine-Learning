```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=300,noise=.1)
x_outliers=np.random.uniform(low=-3,high=3,size=(20,2))
```

```python
x.shape
```

```python
x=np.vstack((x,x_outliers))
x.shape
```

```python
lof=LocalOutlierFactor(n_neighbors=20,contamination=.1)
y_pred=lof.fit_predict(x)
```

```python
y_pred
```

```python
x_inliers=x[y_pred==1]
x_outliers=x[y_pred==-1]
```

```python
x_inliers # all values of x where y is 1
```

```python
plt.title("lof on moons data")
plt.scatter(x_inliers[:,0],x_inliers[:,1],s=20,label='normal data point')
plt.scatter(x_outliers[:,0],x_outliers[:,1],s=20,label='outliers')
plt.show()
```

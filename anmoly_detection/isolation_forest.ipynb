```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('healthcare.csv')
```

```python
plt.scatter(x=df['0'], y=df['1'], c='pink')
plt.show()
```

```python
#plt.scatter(df.iloc[:,0:1],df.iloc[:,1:2]) also correct
plt.scatter(x=df.iloc[:,0],y=df.iloc[:,1])
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
isf=IsolationForest(contamination=0.2) #Expected proportion of outliers (anomalies) in the dataset.
isf.fit(df)
```

```python
pred=isf.predict(df)
pred
```

```python
pred=np.where(pred ==-1,0,pred)
pred
```

```python
indices=np.where(pred==0)
print(indices)
```

```python
indices=np.where(pred==0)[0]
print(indices)
```

```python
plt.scatter(df.iloc[indices,0:1],df.iloc[indices,1:2])
plt.show()
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
plt.scatter(df.iloc[indices, 0],
            df.iloc[indices, 1],edgecolors='orange')

plt.show()
```

```python
X = df.values

plt.scatter(X[indices,0], X[indices,1])
plt.show()
```

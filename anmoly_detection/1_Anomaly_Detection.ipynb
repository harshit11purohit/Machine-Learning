````python
# What is Anomaly Detection?
# Anomaly detection, also known as outlier detection, is the process of identifying data points that deviate significantly from the majority of the data. These anomalous data points can indicate critical incidents, such as fraud, network intrusions, or equipment failures. Anomaly detection is crucial in various fields, including finance, cybersecurity, healthcare, and manufacturing.

# In simple terms, anomaly detection is like finding a needle in a haystack. It involves analyzing data to identify patterns and then flagging any data points that do not conform to those patterns as anomalies.

# Anomaly could be an error, fraud, or a rare event. For example, in a credit card transaction dataset, an anomaly could be a transaction that is significantly larger than the typical transactions made by a user, which could indicate fraudulent activity.

# Everyday Examples
# 1. Bank Transaction Monitoring:
# You normally spend ₹2000–₹5000 per week.
# One day, you make a transaction of ₹50,000. This transaction would be flagged as an anomaly because it deviates significantly from your usual spending pattern.
# 2. Network Security:
# A user typically logs in from a specific location and device.
# If there is a login attempt from a different location or device, it could be flagged as an anomaly, potentially indicating a security breach.
# 3. Healthcare:
# A patient’s heart rate is usually between 60 and 100 beats per minute.
# If the heart rate suddenly spikes to 150 beats per minute, it could be flagged as an anomaly, indicating a potential health issue that requires immediate attention.
# 4. Manufacturing:
# A machine typically produces 1000 units per day.
# If the production suddenly drops to 500 units, it could be flagged as an anomaly, indicating a potential issue with the machine that needs to be addressed.
# 5. E-commerce:
# A customer typically purchases items worth ₹1000–₹3000 per month.
# If there is a sudden purchase of ₹20,000, it could be flagged as an anomaly, potentially indicating fraudulent activity or a significant change in the customer’s behavior.
# 6. Social Media:
# A user typically posts 1–2 times per day.
# If there is a sudden increase to 10 posts in a day, it could be flagged as an anomaly, potentially indicating a change in the user’s behavior or a possible security issue with their account.
# 7. Environmental Monitoring:
# A sensor typically records a temperature of 25°C.
# If the sensor suddenly records a temperature of 50°C, it could be flagged as an anomaly, indicating a potential malfunction of the sensor or an extreme environmental condition that needs to be investigated.
# 8. Stock Market:
# A stock typically trades within a certain price range.
# If there is a sudden spike or drop in the stock price, it could be flagged as an anomaly, potentially indicating market manipulation, a significant news event, or a change in investor sentiment that needs to be analyzed further.
# 9. Email Spam Detection:
# A user typically receives a certain number of emails per day.
# If there is a sudden influx of emails, especially from unknown senders, it could be flagged as an anomaly, potentially indicating a spam attack or a phishing attempt that needs to be addressed to protect the user’s information and security.
# 10. Energy Consumption:
# A household typically consumes a certain amount of electricity each month.
# If there is a sudden spike in energy consumption, it could be flagged as an anomaly, potentially indicating a malfunctioning appliance, a change in the household’s behavior, or an issue with the energy meter that needs to be investigated to prevent unnecessary costs and ensure efficient energy usage.
````

````python
# Types of Anomalies and Challenges
# Anomalies can be grouped into three main types, based on how they differ from normal behavior.

# 1. Point Anomaly (Most Common)
# A single data point is anomalous compared to the rest of the data.
# Example: Credit card transaction
# In a credit card transaction dataset, a point anomaly could be a transaction of ₹50,000 when the user typically spends between ₹2000 and ₹5000. This single transaction stands out as an anomaly because it deviates significantly from the user’s normal spending pattern.
# Use cases:
# - Fraud detection in financial transactions
# - Intrusion detection in network security
# - Fault detection in manufacturing processes

# 2. Contextual Anomaly (aka Conditional Anomaly)
# A data point is anomalous in a specific context but may be normal in another context.
# Example: Temperature readings
# A temperature reading of 39°C is an anomaly when it occurs in the human body, indicating a fever. However, the same temperature reading of 39°C is normal when it occurs in the summer in Delhi. This illustrates how the same value can be an anomaly in one context (human body) and normal in another context (summer weather), highlighting the importance of considering the context when identifying anomalies.
# Use cases:
# - Anomaly detection in time-series data (e.g., sensor readings)
# - Anomaly detection in spatial data (e.g., geographic data)
# - Anomaly detection in user behavior (e.g., login patterns)

# 3. Collective Anomaly (aka Group Anomaly)
# A collection of data points is anomalous together, even if individual points may not be anomalous on their own.
# Example: Network traffic
# In network traffic analysis, a collective anomaly could be a sudden surge in traffic from multiple IP addresses to a specific server. Individually, each IP address may not show anomalous behavior, but when considered together, the surge in traffic indicates a potential Distributed Denial of Service (DDoS) attack, which is a collective anomaly.
# Use cases:
# - Anomaly detection in network traffic (e.g., DDoS attacks)
# - Anomaly detection in social networks (e.g., coordinated fake accounts)
# - Anomaly detection in financial markets (e.g., coordinated trading activity)
````

````python
# Key Challenges in Anomaly Detection
# Even though anomaly detection sounds simple, it is very challenging in practice.

# 1. Lack of Labeled Anomaly Data
# Anomalies are rare, so there is often a lack of labeled data for training machine learning models. This makes it difficult to apply supervised learning techniques for anomaly detection, as there may not be enough examples of anomalies to learn from.
# eg: In a credit card fraud detection dataset, there may be only a small percentage of transactions that are labeled as fraudulent (anomalies), while the majority of transactions are labeled as normal. This imbalance in the dataset can make it challenging to train a machine learning model effectively, as the model may struggle to learn the patterns associated with the rare anomalies.

# 2. High Dimensionality
# Anomaly detection often involves analyzing high-dimensional data, which can make it difficult to identify anomalies effectively. High-dimensional data can lead to the "curse of dimensionality," where the distance between data points becomes less meaningful, making it harder to distinguish between normal and anomalous data points. Additionally, high-dimensional data may contain irrelevant features that can introduce noise and further complicate the anomaly detection process.
# eg: In a network intrusion detection dataset, there may be hundreds of features representing various aspects of network traffic. The high dimensionality of the data can make it challenging to identify anomalies, as the model may struggle to find meaningful patterns in the data and may be overwhelmed by irrelevant features that do not contribute to anomaly detection.

# 3. Evolving Anomalies
# Anomalies can evolve over time, making it difficult to maintain an effective anomaly detection system. As new types of anomalies emerge, the patterns associated with them may change, requiring continuous updates to the anomaly detection model. This is particularly challenging in dynamic environments, such as cybersecurity, where attackers constantly adapt their techniques to evade detection. Anomaly detection systems must be able to adapt to these changes in order to remain effective in identifying new and evolving anomalies.
# eg: In cybersecurity, attackers may develop new methods to bypass existing security measures, leading to the emergence of new types of anomalies in network traffic. An anomaly detection system that was trained on historical data may not be able to identify these new anomalies effectively, as the patterns associated with them may differ from previously observed anomalies. This highlights the need for anomaly detection systems to be adaptable and continuously updated to keep up with evolving threats and anomalies in dynamic environments.

# 4. Imbalanced Data
# Anomalies are often rare compared to normal data, leading to imbalanced datasets. This imbalance can make it difficult for machine learning models to learn the patterns associated with anomalies, as they may be overwhelmed by the majority class (normal data). As a result, models may have a tendency to predict the majority class, leading to poor performance in detecting anomalies. Techniques such as resampling, anomaly detection algorithms that are designed to handle imbalanced data, and evaluation metrics that account for class imbalance can be used to address this challenge.
# eg: In a credit card fraud detection dataset, there may be only a small percentage of transactions that are labeled as fraudulent (anomalies), while the majority of transactions are labeled as normal. This imbalance in the dataset can lead to a situation where a machine learning model may predict most transactions as normal, resulting in a high accuracy but poor performance in detecting fraudulent transactions. To address this challenge, techniques such as oversampling the minority class (fraudulent transactions) or using anomaly detection algorithms that are designed to handle imbalanced data can be employed to improve the model's ability to detect anomalies effectively. Additionally, evaluation metrics such as precision, recall, and F1-score can be used to assess the model's performance in detecting anomalies, rather than relying solely on accuracy, which may be misleading in imbalanced datasets.

# 5. Noise and Outliers
# Anomalies can be confused with noise or outliers in the data, making it difficult to distinguish between them. Noise refers to random variations in the data that do not represent meaningful patterns, while outliers are data points that deviate significantly from the majority of the data but may not necessarily be anomalies. Distinguishing between true anomalies and noise or outliers can be challenging, as they may share similar characteristics. It is important to carefully analyze the data and use appropriate techniques to differentiate between anomalies, noise, and outliers in order to ensure accurate anomaly detection.
# eg: In a sensor dataset, there may be occasional spikes in the readings due to random fluctuations (noise) or malfunctioning sensors (outliers). These spikes may not necessarily indicate true anomalies, but they can be mistaken for anomalies if not properly analyzed. It is important to use techniques such as data preprocessing, statistical analysis, and domain knowledge to differentiate between true anomalies and noise or outliers in order to ensure accurate anomaly detection and avoid false positives.

# 6. Dynamic and Streaming Data
# Anomaly detection in dynamic and streaming data can be particularly challenging, as the data is continuously changing and evolving over time. In such scenarios, traditional batch processing methods may not be suitable for real-time anomaly detection. Instead, online or incremental learning algorithms that can adapt to changing data distributions and detect anomalies in real-time are required. Additionally, handling concept drift, where the underlying patterns in the data change over time, can further complicate anomaly detection in dynamic and streaming data. It is important to implement anomaly detection systems that can effectively handle dynamic and streaming data to ensure timely detection of anomalies in real-world applications.
# eg: In a network traffic monitoring system, the data is continuously generated as users interact with the network. Anomaly detection in this context requires real-time analysis of the streaming data to identify potential security threats or unusual patterns in network traffic. Traditional batch processing methods may not be suitable for this scenario, as they may introduce delays in detecting anomalies. Instead, online or incremental learning algorithms that can adapt to changing data distributions and detect anomalies in real-time are necessary to ensure timely detection of potential security threats or unusual patterns in network traffic. Additionally, handling concept drift, where the underlying patterns in the network traffic may change over time, is crucial for maintaining the effectiveness of the anomaly detection system in a dynamic and streaming data environment.
````

````python
# Detection Approaches – Supervised, Unsupervised, Semi-Supervised, and Self-Supervised Learning

# 1. Supervised Anomaly Detection
# Description:
# In supervised anomaly detection, the model is trained on a labeled dataset that contains both normal and anomalous instances. 
# The model learns to distinguish between the two classes based on the features of the data. This approach is effective when there 
# is a sufficient amount of labeled data available for training. However, it can be challenging to obtain labeled data for anomalies
# , as they are often rare and may require manual labeling by experts.
# Use cases:
# - Fraud detection in financial transactions
# - Intrusion detection in network security
# - Fault detection in manufacturing processes
# Algorithms:
# - Decision Trees
# - Random Forests
# - Support Vector Machines (SVM)
# - Neural Networks

# 2. Unsupervised Anomaly Detection
# Description:
# In unsupervised anomaly detection, the model is trained on an unlabeled 
# dataset that contains only normal instances. The model learns the
# underlying patterns of the normal data and identifies anomalies as
# data points that deviate significantly from these patterns. 
# This approach is useful when labeled data is not available, but it 
# can be less accurate than supervised methods, as it relies on the
# assumption that anomalies are rare and significantly different from normal
# data.
# Use cases:
# - Anomaly detection in time-series data (e.g., sensor readings)
# - Anomaly detection in spatial data (e.g., geographic data)
# - Anomaly detection in user behavior (e.g., login patterns)
# Algorithms:
# - Clustering algorithms (e.g., K-Means, DBSCAN)
# - Density-based methods (e.g., Local Outlier Factor)
# - Autoencoders
# - Isolation Forest

# 3. Semi-Supervised Anomaly Detection
# Description:
# In semi-supervised anomaly detection, the model is trained on a dataset that contains a small amount of labeled data (both normal and anomalous) and a larger amount of unlabeled data. The model learns from the labeled data to identify patterns of normal and anomalous instances, and then applies this knowledge to the unlabeled data to detect anomalies. This approach can be effective when there is a limited amount of labeled data available, but it still requires some labeled examples of anomalies for training.
# Use cases:
# - Fraud detection in financial transactions with limited labeled data
# - Intrusion detection in network security with limited labeled data
# - Fault detection in manufacturing processes with limited labeled data
# Algorithms:
# - Semi-supervised SVM
# - Semi-supervised Neural Networks
# - Semi-supervised Clustering
# - Semi-supervised Autoencoders
````

````python
# Isolation Forest Python Implementation
!pip install spacy
import pandas as pd
import random
import spacy

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Step 1: Normal queries (25 entries)
normal_queries = [
    "How to learn Python",
    "Best pizza places in New York",
    "Current weather in London",
    "Top 10 movies of 2023",
    "How to cook spaghetti",
    "Latest news in technology",
    "Easy yoga poses for beginners",
    "Best laptops under 1000 dollars",
    "Famous paintings by Van Gogh",
    "Upcoming Marvel movies",
    "Top universities in the world",
    "What is quantum computing",
    "Symptoms of flu vs COVID",
    "How to bake chocolate cake",
    "Best books to read in 2024",
    "Tips for saving money fast",
    "Who won the World Cup 2022?",
    "History of the Roman Empire",
    "Popular tourist spots in Japan",
    "How to start a blog",
    "Ways to improve communication skills",
    "Best exercises to lose belly fat",
    "Current stock market trends",
    "What is artificial intelligence?",
    "Famous quotes by Albert Einstein"
]

# Step 2: Anomalous queries (5 entries)
anomalous_queries = [
    "Python",                           # too short
    "asdf qwerty zxcv",                 # gibberish
    "Open file system root access now",# command-style
    "weatherweatherweatherweather",    # repeated nonsense
    "How much is is is is this??"      # repetitive and strange
]

# Combine and label
queries = normal_queries + anomalous_queries
labels = [0]*len(normal_queries) + [1]*len(anomalous_queries)

# Step 3: Feature extraction
data = []

for query, label in zip(queries, labels):
    doc = nlp(query)
    query_length = len(query)
    word_count = len(query.split())
    noun_count = sum(1 for token in doc if token.pos_ in ['NOUN', 'PROPN'])
    is_question = int(query.strip().endswith("?") or query.lower().startswith(("how", "what", "why", "who", "where", "is", "are", "can")))

    data.append({
        "Query": query,
        "Query Length": query_length,
        "Word Count": word_count,
        "Noun Count": noun_count,
        "Is Question": is_question,
        "Anomaly": label
    })

# Step 4: Save to CSV
df = pd.DataFrame(data)
df.to_csv("query_anomaly_dataset.csv", index=False)
print("Dataset saved as query_anomaly_dataset.csv")
df
````

````python
import pandas as pd

# Load the dataset
df = pd.read_csv("query_anomaly_dataset.csv")

# Drop non-numeric (text) columns for Isolation Forest
features = df[["Query Length", "Word Count", "Noun Count", "Is Question"]]

# (Optional) See the data
features.head()

````

````python
# Goal: Use a machine learning model to detect anomalous queries without supervision (unsupervised anomaly detection)

# Apply Isolation Forest
# What is Isolation Forest?
# Isolation Forest is an unsupervised anomaly detection algorithm that works by isolating anomalies instead of profiling normal data points. It builds an ensemble of trees (isolation trees) where anomalies are more likely to be isolated in fewer splits compared to normal data points. The algorithm assigns an anomaly score to each data point, with higher scores indicating a higher likelihood of being an anomaly. Isolation Forest is effective for high-dimensional datasets and can handle large datasets efficiently, making it a popular choice for anomaly detection tasks.
from sklearn.ensemble import IsolationForest
# Create the Isolation Forest model
iso_forest = IsolationForest(n_estimators=100, contamination=0.15, random_state=42)
# n_estimators: The number of trees in the forest. A higher number of trees can improve the model's performance but also increases computational cost.
# contamination: The proportion of anomalies in the dataset. This parameter helps the model to determine the threshold for classifying data points as anomalies. In this case, we set it to 0.15, which means we expect 15% of the data to be anomalies. Adjusting this parameter can help improve the model's performance based on the expected proportion of anomalies in the dataset.

df["anomaly_score"] = iso_forest.fit_predict(features)
# fit_predict returns -1 for anomalies and 1 for normal data points. We can use this to create an "anomaly_score" column in the DataFrame, where -1 indicates an anomaly and 1 indicates a normal query.

df.head()
# anomaly_score will be:
# -1: anomaly (outlier)
# 1: normal data point
````

````python
# Evaluate Results
# Map Isolation Forest output
# anomaly_score column contains:
# 1 → Normal
# -1 → Anomaly

# But your true labels are in the "Anomaly" column:
# 0 → Normal
# 1 → Anomaly
df["predicted_label"] = df["anomaly_score"].map({1: 0, -1: 1})

# Compare with actual labels (if available)
if "Anomaly" in df.columns:
    from sklearn.metrics import classification_report, confusion_matrix

    print("Confusion Matrix:")
    print(confusion_matrix(df["Anomaly"], df["predicted_label"]))

    print("\nClassification Report:")
    print(classification_report(df["Anomaly"], df["predicted_label"]))
else:
    print("True labels not available for evaluation.")
````

````python
# Local Outlier Factor (LOF)
# What is Local Outlier Factor (LOF)?
# Local Outlier Factor (LOF) is an unsupervised anomaly detection algorithm that identifies anomalies based on the local density of data points. It calculates the local density of a data point by comparing it to the densities of its neighbors. A data point is considered an anomaly if its local density is significantly lower than that of its neighbors. LOF is effective in detecting anomalies in datasets with varying densities and can handle high-dimensional data. It assigns an anomaly score to each data point, with higher scores indicating a higher likelihood of being an anomaly. LOF is particularly useful for identifying anomalies in complex datasets where the density of normal data points may vary across different regions of the feature space.

import pandas as pd
from sklearn.neighbors import LocalOutlierFactor

# Load dataset
df = pd.read_csv("query_anomaly_dataset.csv")

# Use only numeric features for LOF
features = df[["Query Length", "Word Count", "Noun Count", "Is Question"]]
features.head(10)
````

````python
 # Goal:You already prepared a feature matrix (features) from query data. Now you're applying LOF to detect
#  which queries are anomalous based on how isolated they are compared to their local neighbors.

# Apply Local Outlier Factor
# Initialize LOF model
lof = LocalOutlierFactor(n_neighbors=5, contamination=0.15)
# n_neighbors: The number of neighbors to use for calculating the local density. A smaller value may capture more local anomalies, while a larger value may capture more global anomalies. In this case, we set it to 5, which means the algorithm will consider the 5 nearest neighbors when calculating the local density for each data point.
# contamination: The proportion of anomalies in the dataset. This parameter helps the model to determine the threshold for classifying data points as anomalies. In this case, we set it to 0.15, which means we expect 15% of the data to be anomalies. Adjusting this parameter can help improve the model's performance based on the expected proportion of anomalies in the dataset.

# LOF internally:
# Calculates local reachability density (LRD) for each point
# Compares LRD to its neighbors
# Anomalies have significantly lower LRD → lower LOF score → flagged as anomalies

# Predict: -1 = outlier, 1 = inlier
df["lof_score"] = lof.fit_predict(features)

# Convert to 0 = normal, 1 = anomaly
df["predicted_label_lof"] = df["lof_score"].map({1: 0, -1: 1})
df.head()
````

````python
# Evaluate (If True Labels Exist)
from sklearn.metrics import confusion_matrix, classification_report

if "Anomaly" in df.columns:
    print("Confusion Matrix:")
    print(confusion_matrix(df["Anomaly"], df["predicted_label_lof"]))

    print("\nClassification Report:")
    print(classification_report(df["Anomaly"], df["predicted_label_lof"]))
else:
    print("True labels not available for evaluation.")
````

````python
# Convert confusion matrix to DataFrame for better visualization
conf_matrix_lof = confusion_matrix(df["Anomaly"], df["predicted_label_lof"])
conf_matrix_lof_df = pd.DataFrame(conf_matrix_lof, index=["Actual Normal", "Actual Anomaly"], columns=["Predicted Normal", "Predicted Anomaly"])
print("Confusion Matrix (LOF) as DataFrame:")
conf_matrix_lof_df
````

````python
# One-Class SVM
# Learns a boundary around the majority class (normal data).
# Good when you have mostly normal data and very few anomalies.

# Why use it?
# Works well for high-dimensional, small datasets.
# Can model non-linear boundaries (with kernels).
````

````python
# Import Required Libraries
import pandas as pd
from sklearn.svm import OneClassSVM
from sklearn.metrics import classification_report, confusion_matrix
````

````python
# Load the Dataset
# Load the dataset
df = pd.read_csv("query_anomaly_dataset.csv")

# Select features (must be numerical)
features = df[["Query Length", "Word Count", "Noun Count", "Is Question"]]
features
````

````python
# Use One-Class SVM to:
# Learn the “normal query” patterns from features like query length, word count, etc.
# Flag anything that doesn’t fit this pattern as an anomaly

# Fit One-Class SVM

# Initialize One-Class SVM
# nu = expected proportion of anomalies (set to 0.15 or tweak)
oc_svm = OneClassSVM(kernel='rbf', gamma='auto', nu=0.15)
# kernel='rbf': Uses the Radial Basis Function kernel, which allows the model to capture non-linear relationships in the data. This is useful for anomaly detection, as anomalies may not always be linearly separable from normal data points. Others include 'linear', 'poly', etc.
# gamma='auto': Sets the gamma parameter to 1/n_features, which is a common default value that can work well in many cases. Gamma controls the influence of individual data points on the decision boundary. A smaller gamma value means that the model will consider more data points when making decisions, while a larger gamma value means that the model will focus more on individual data points. Setting it to 'auto' allows the model to automatically adjust gamma based on the number of features in the dataset.
# nu=0.15: Sets the nu parameter, which is an upper bound on the fraction of anomalies in the dataset. In this case, we set it to 0.15, which means we expect 15% of the data to be anomalies. Adjusting this parameter can help improve the model's performance based on the expected proportion of anomalies in the dataset.

# This tells the model to enclose ~85% of the points and reject the rest as anomalies.

# Fit and predict
df["svm_score"] = oc_svm.fit_predict(features)

# Convert output: -1 = anomaly → 1, 1 = normal → 0
df["predicted_label_svm"] = df["svm_score"].map({1: 0, -1: 1})
df
````

````python
# DBSCAN (Density-Based Spatial Clustering of Applications with Noise) for anomaly detection
# What is DBSCAN?
# DBSCAN is a density-based clustering algorithm that groups together data points that are closely packed together while marking points that lie alone in low-density regions as outliers (anomalies). It requires two parameters: eps (the maximum distance between two samples for them to be considered as in the same neighborhood) and min_samples (the number of samples in a neighborhood for a point to be considered as a core point). DBSCAN is effective for anomaly detection because it can identify clusters of normal data points and flag points that do not belong to any cluster as anomalies. It is particularly useful for datasets with noise and can handle clusters of varying shapes and sizes.

# Import Libraries
import pandas as pd
from sklearn.cluster import DBSCAN # DBSCAN is a density-based clustering algorithm that can be used for anomaly detection by identifying points that do not belong to any cluster as anomalies.
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
````

````python
# Load the Dataset
# Load the dataset
df = pd.read_csv("query_anomaly_dataset.csv")

# Use only numeric features
features = df[["Query Length", "Word Count", "Noun Count", "Is Question"]]
features
````

````python
# Apply DBSCAN
# You want to cluster normal queries and flag isolated or sparse queries as anomalies using DBSCAN.

# Initialize DBSCAN
dbscan = DBSCAN(eps=3, min_samples=3)
# eps: The maximum distance between two samples for them to be considered as in the same neighborhood. A smaller eps value will result in more clusters and potentially more anomalies, while a larger eps value may result in fewer clusters and fewer anomalies. In this case, we set it to 3, which means that data points within a distance of 3 from each other will be considered part of the same cluster.
# Ideal eps can be found using a k-distance graph, where you plot the distance to the k-th nearest neighbor for each point and look for the "elbow" point.
# min_samples: The number of samples in a neighborhood for a point to be considered as a core point. A smaller min_samples value will result in more clusters and potentially more anomalies, while a larger min_samples value may result in fewer clusters and fewer anomalies. In this case, we set it to 3, which means that a point must have at least 3 neighbors within the eps distance to be considered a core point. Points that do not meet this criterion will be considered as noise (anomalies).

# Fit DBSCAN and assign cluster labels
dbscan_labels = dbscan.fit_predict(features)

# Add cluster labels to DataFrame
df["dbscan_label"] = dbscan_labels

# Mark anomalies: DBSCAN assigns -1 to outliers
df["predicted_label_dbscan"] = (dbscan_labels == -1).astype(int)
# True → 1 (anomaly)
# False → 0 (normal)
df.head()
````

````python
# Lets find the optimal eps value using a k-distance graph
from sklearn.neighbors import NearestNeighbors
# Calculate the distance to the 3rd nearest neighbor for each point
neighbors = NearestNeighbors(n_neighbors=3)
neighbors_fit = neighbors.fit(features)
distances, indices = neighbors_fit.kneighbors(features)
# Sort the distances to the 3rd nearest neighbor
distances = distances[:, 2]  # Get the distance to the 3rd nearest neighbor
distances.sort()
# Plot the k-distance graph
plt.figure(figsize=(10, 6))
plt.plot(distances)
plt.xlabel("Data Points sorted by distance to 3rd nearest neighbor")
plt.ylabel("Distance to 3rd Nearest Neighbor")
plt.title("K-Distance Graph for DBSCAN")
plt.axhline(y=3, color='r', linestyle='--')  # Add a horizontal line at the chosen eps value
plt.show()
````

**Interpretation of the K-Distance Graph for DBSCAN**

The purpose of the **K-Distance Graph** is to determine the optimal value of the **`eps` (epsilon)** parameter for the DBSCAN clustering algorithm.

Step-by-step interpretation of your graph:

1. **What the graph represents**

   * The **X-axis** represents the data points sorted by their distance to the **3rd nearest neighbor**.
   * The **Y-axis** represents the **distance to the 3rd nearest neighbor** for each data point.
   * Since `n_neighbors=3` was used, DBSCAN will consider the **distance to the 3rd nearest neighbor** when deciding cluster density.

2. **Understanding the curve**

   In the graph, the curve shows three regions:

   **Region 1 — Flat region (Dense cluster area)**
   From the beginning up to around point ~24–25, the distance values increase **very slowly** (around 1 to about 2.5).
   This means:

   * Points are **close to their neighbors**
   * These points belong to **dense regions**
   * They are **likely part of clusters**

   **Region 2 — Elbow / Knee point**
   Around **distance ≈ 3**, the curve starts increasing sharply.

   This point is called the **knee** or **elbow** of the graph.

   This is the **ideal location to choose `eps`**.

   **Region 3 — Sharp rise (Outliers or sparse points)**
   After the elbow, the distances increase very quickly (reaching up to ~13).

   These points:

   * Have large distances to neighbors
   * Are located in **sparse regions**
   * Will likely be detected as **noise or outliers**

3. **Why `eps ≈ 3` is a good choice**

   The horizontal red line in the graph is placed at:

   ```
   eps = 3
   ```

   This value is appropriate because:

   * It lies at the **transition point between dense and sparse regions**
   * Points below this threshold form **clusters**
   * Points above this threshold become **noise**

4. **Expected DBSCAN behavior with `eps = 3`**

   If you run DBSCAN with:

   ```
   eps = 3
   min_samples = 3
   ```

   then the algorithm will:

   * Identify dense clusters among the first ~25 points
   * Mark the last few points (with very large distances) as **outliers**
   * Produce clusters that reflect the natural density structure of the data

5. **Important practical insight**

   The **key idea of the k-distance graph** is:

   The optimal `eps` is usually chosen at the **knee point where the curve starts rising sharply**.

   In your graph, that transition happens around:

   ```
   eps ≈ 3
   ```

   Therefore, the chosen value in your code is **correct and justified**.

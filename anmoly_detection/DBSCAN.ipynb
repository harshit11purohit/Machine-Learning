```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
```

```python
x,y=make_circles(n_samples=800,factor=.3,noise=.1)
x
```

```python
df=pd.DataFrame(x,columns=[0,1])
df
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
plt.show()
```

```python
db=DBSCAN(eps=0.1,min_samples=10)
out=db.fit_predict(x)
out.shape
```

```python
db
```

```python
db.labels_
```

```python
out
```

```python
plt.scatter(x[:,0],x[:,1],c=db.labels_)
plt.show()
```

```python
plt.scatter(x[out,0],x[out,1])
plt.show()
```

```python
plt.scatter(x[out == -1, 0],
            x[out == -1, 1])

plt.show()
```

```python
# clusters
plt.scatter(x[out != -1, 0],
            x[out != -1, 1],
            c=out[out != -1],
            label="Clusters")

# noise points
plt.scatter(x[out == -1, 0],
            x[out == -1, 1],
            color='red',
            label="Noise")

plt.legend()
plt.show()
```

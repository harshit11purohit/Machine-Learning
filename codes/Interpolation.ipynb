```python
'''Interpolation TheoryDefinition: Interpolation is the mathematical process of estimating unknown values that fall between known data points.Linear vs. Cubic: Linear interpolation connects points with straight lines ($y = mx + c$), while Cubic Spline uses third-degree polynomials to create a smooth, continuous curve.Use Case: Essential in Machine Learning for handling missing data, increasing signal resolution, or smoothing noisy sensor inputs.Overfitting Risk: High-degree polynomial interpolation can lead to Runge's Phenomenon, where the curve oscillates wildly at the edges of the interval.'''
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Original data points (with some missing in between)
x = np.array([1, 3, 5, 7])
y = np.array([10, 30, 50, 70])

# Generate smooth x values using linspace for plotting the line
x_line = np.linspace(x.min(), x.max(), 100)
x_line
# Linear interpolation using np.interp

```

```python
y_line = np.interp(x_line, x, y)
y_line
# Plot
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x, y, color='red', label='Original Data')
plt.plot(x_line, y_line, color='blue', label='Linear Interpolation')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Interpolation with Scatter and Line")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np

# Sample data with missing values
data = {
    'Day': [1, 2, 3, 4, 5],
    'Sales': [100, np.nan, 300, np.nan, 500]
}

df = pd.DataFrame(data)

# Linear interpolation
df['Sales'] = df['Sales'].interpolate(method='linear')

print(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Original data points
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 20, 15, 25, 30])

# Cubic interpolation function
cubic_interp = interp1d(x, y, kind='cubic')

# Generate smooth x values for plotting
x_smooth = np.linspace(x.min(), x.max(), 100)
y_smooth = cubic_interp(x_smooth)

# Plot
plt.figure(figsize=(8,5))
plt.scatter(x, y, color='red', label='Original Data')
plt.plot(x_smooth, y_smooth, color='blue', label='Cubic Interpolation')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Cubic Interpolation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Original data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 6, 5, 10, 8])

# Fit polynomial of degree 2
coeffs = np.polyfit(x, y, deg=2)  # returns [a2, a1, a0]
print("Coefficients:", coeffs)

# Create polynomial function
poly = np.poly1d(coeffs)

# Generate smooth x for plotting
x_smooth = np.linspace(1, 5, 100)
y_smooth = poly(x_smooth)

# Plot
plt.scatter(x, y, color='red', label='Original Data')
plt.plot(x_smooth, y_smooth, color='blue', label='Polynomial Fit (deg=2)')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial Regression using np.polyfit")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data
np.random.seed(1)
data = np.random.normal(50, 10, 100)
data = np.append(data, [150, 160])  # adding outliers

df = pd.DataFrame(data, columns=['Value'])

# Box plot
plt.figure(figsize=(6,4))
plt.boxplot(df['Value'], vert=True, patch_artist=True)
plt.title("Box Plot to Detect Outliers")
plt.ylabel("Value")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(6,4))
sns.histplot(df['Value'], bins=20, kde=True)
plt.title("Distribution Plot to Check Outliers")
plt.xlabel("Value")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

x = np.linspace(0, 10, num=11)
y = np.cos(-x**2/9.0)

x_new = np.linspace(0, 10, num=100)

f_linear = interp1d(x, y, kind='linear')
f_cubic = interp1d(x, y, kind='cubic')

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o', label='Data Points')
plt.plot(x_new, f_linear(x_new), '-', label='Linear')
plt.plot(x_new, f_cubic(x_new), '--', label='Cubic Spline')
plt.legend()
plt.show()
```

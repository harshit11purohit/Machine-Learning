```python
import pandas as pd
import numpy as np

np.random.seed(1)

no_of_samples = 1000
class_0_ratio = 0.9
class_1_ratio = 0.1

no_class_0 = int(no_of_samples * class_0_ratio)
no_class_1 = int(no_of_samples * class_1_ratio)

class_0 = {
    'feature1': np.random.normal(0, 1, no_class_0),
    'feature2': np.random.normal(0, 1, no_class_0),
    'target': [0] * no_class_0
}

class_1 = {
    'feature1': np.random.normal(3, 1, no_class_1),
    'feature2': np.random.normal(3, 1, no_class_1),
    'target': [1] * no_class_1
}

x = pd.DataFrame(class_0)
y = pd.DataFrame(class_1)

df = pd.concat([x, y], ignore_index=True)

df['target'].value_counts()

```

```python
df_majority = df[df['target'] == 0]
df_minority = df[df['target'] == 1]

df_majority.shape   # (900, 3)
df_minority.shape   # (100, 3)
```

```python
from sklearn.utils import resample

df_minority_upsampled = resample(
    df_minority,
    replace=True,
    n_samples=len(df_majority),
    random_state=1
)
```

```python
df_majority_downsampled = resample(
    df_majority,
    replace=False,
    n_samples=len(df_minority),
    random_state=1
)
```

```python
df_majority_downsampled
```

```python
df_downsampled = pd.concat([df_minority, df_majority_downsampled])

df_downsampled['target'].value_counts()
```

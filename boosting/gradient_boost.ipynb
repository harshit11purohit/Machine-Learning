```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.datasets import make_regression
from sklearn.tree import DecisionTreeRegressor
```

```python
x,y=make_regression(n_samples=1000,n_features=4,random_state=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=1)
gbr=GradientBoostingRegressor()
gbr.fit(x_train,y_train)
y_pred=gbr.predict(x_test)
print(mean_squared_error(y_test,y_pred)),
print(r2_score(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
gdr=GradientBoostingRegressor()
param_grid = {
    'n_estimators': [10,20,50,100,200],
    'learning_rate': [0.01, 0.1, 1.0, 1.5, 2.0],
    'max_depth':[1,2,3,4,5]
    }
clf=GridSearchCV(estimator=gdr,param_grid=param_grid,cv=3,verbose=3)
clf
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
print(mean_squared_error(y_test,y_pred)),
print(r2_score(y_test,y_pred))
```

```python
from sklearn.model_selection import RandomizedSearchCV
gdr=GradientBoostingRegressor()
param_ = {
    'n_estimators': [100, 200,300],
    'learning_rate': [0.01, 0.1, 1.0, 1.5, 2.0],
    #algorithm': ['SAMME']
}
clf=RandomizedSearchCV(estimator=gdr,param_distributions=param_,cv=3,verbose=3)
clf
clf.fit(x_train,y_train)
y_pred=clf.best_estimator_.predict(x_test)
print(mean_squared_error(y_test,y_pred)),
print(r2_score(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

ada=GradientBoostingClassifier()
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1.0, 1.5, 2.0],
    #algorithm': ['SAMME']
}
x,y=make_classification(n_classes=2,n_samples=1000,n_features=20,random_state=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=1)
clf=GridSearchCV(estimator=ada,param_grid=param_grid,cv=5,verbose=3)
clf.fit(x_train,y_train)
y_pred=clf.best_estimator_.predict(x_test)
print(accuracy_score(y_test,y_pred)),
print(confusion_matrix(y_test,y_pred)),
print(classification_report(y_test,y_pred))
```

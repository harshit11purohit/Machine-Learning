```python
!pip install catboost
import sys
!{sys.executable} -m pip install catboost
from catboost import CatBoostClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

x, y = make_classification(n_classes=2, n_samples=1000, n_features=20, random_state=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

cbc = CatBoostClassifier(verbose=0)

param_grid = {
    'iterations': [10,20,50,100],
    'learning_rate': [0.01, 0.1, 0.2],
    'depth': [1,2,3,4],
    'l2_leaf_reg': [1, 3, 5]
}

clf = GridSearchCV(estimator=cbc, param_grid=param_grid, cv=5, verbose=3)
clf.fit(x_train, y_train)

y_pred = clf.best_estimator_.predict(x_test)

print(accuracy_score(y_test, y_pred)),
print(confusion_matrix(y_test, y_pred)),
print(classification_report(y_test, y_pred))
```

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.datasets import make_regression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

x, y = make_regression(n_samples=1000, n_features=20, noise=0.1, random_state=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

cbr = CatBoostRegressor(verbose=0)

param_distributions = {
    'iterations': [50, 100,10],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'depth': [4,0,1,2,3],
    'l2_leaf_reg': [1, 3, 5, 7, 9]
}

search = RandomizedSearchCV(estimator=cbr, param_distributions=param_distributions, n_iter=10, cv=5, verbose=3, random_state=1)
search.fit(x_train, y_train)

y_pred = search.best_estimator_.predict(x_test)

print(mean_squared_error(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))
print(r2_score(y_test, y_pred))
```

```python
# SUMMARY OF MODEL BEHAVIOR:
# 1. 'iterations' (n_estimators): Total number of trees. Too many leads to overfitting; too few leads to underfitting.
# 2. 'learning_rate' (eta): The step size. Smaller values (0.01) require more iterations but usually generalize better.
# 3. 'depth': The complexity of individual trees. CatBoost works well with shallow trees (depth 4-6).
# 4. 'l2_leaf_reg': The L2 regularization coefficient. Increasing this value makes the model more conservative.
# 5. 'verbose=0': Silences the training logs for a cleaner output during search iterations.
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score, classification_report

# Create dataset
X, y = make_classification(
    n_samples=1000,
    n_features=20,
    n_classes=2,
    random_state=42
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Base Model
xgb = XGBClassifier(
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)

# Hyperparameter Grid
param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5, 7],
    'reg_alpha': [0, 0.1, 1],        # L1 regularization
    'reg_lambda': [1, 5, 10],        # L2 regularization
    'gamma': [0, 0.1, 0.5]
}

# Grid Search with Cross Validation
grid = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    verbose=2,
    n_jobs=-1
)

grid.fit(X_train, y_train)

# Best Model
best_model = grid.best_estimator_

# Predictions
y_pred = best_model.predict(X_test)

print("Best Parameters:", grid.best_params_)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# XGBoost is a gradient boosting algorithm that builds trees sequentially,
# where each new tree tries to correct the errors made by previous trees.

# Boosting works by minimizing a loss function using gradient descent
# in function space, making the model progressively improve.

# Hyperparameter tuning is performed using GridSearchCV, which systematically
# explores combinations of parameters to find the configuration that gives
# the best performance.

# Cross-Validation (cv=5) splits the training data into multiple folds and
# trains the model several times to estimate generalization performance
# more reliably and reduce overfitting risk.

# Regularization parameters such as reg_alpha (L1) and reg_lambda (L2)
# control model complexity and help prevent overfitting.

# Gamma defines the minimum loss reduction required to make a split,
# acting as another regularization mechanism.

# Learning rate controls how much each tree contributes to the overall
# model, enabling gradual learning and improved generalization.

# The ensemble of many weak decision trees forms a strong learner
# through additive model building.
```

### Dataset Column Descriptions:

1. **hotel**: Type of hotel where the booking was made (e.g., Resort Hotel or City Hotel).

2. **is_canceled**: Indicates if the booking was canceled (1 = Canceled, 0 = Not Canceled).

3. **lead_time**: Number of days between the booking date and the arrival date.

4. **arrival_date_year**: Year of arrival.

5. **arrival_date_month**: Month of arrival.

6. **arrival_date_week_number**: Week number of the arrival date within the year.

7. **arrival_date_day_of_month**: Day of the month when the guest arrived.

8. **stays_in_weekend_nights**: Number of weekend nights (Friday and Saturday) the guest stayed.

9. **stays_in_week_nights**: Number of weekday nights (Monday to Thursday) the guest stayed.

10. **adults**: Number of adults included in the booking.

11. **children**: Number of children included in the booking.

12. **babies**: Number of babies included in the booking.

13. **meal**: Type of meal plan booked (e.g., BB = Bed & Breakfast, HB = Half Board).

14. **country**: Country of origin of the guest (represented by a country code).

15. **market_segment**: The market segment of the booking (e.g., Online TA, Corporate).

16. **distribution_channel**: How the booking was distributed (e.g., Direct, Corporate, TA/TO).

17. **is_repeated_guest**: Indicates if the guest is a returning customer (1 = Yes, 0 = No).

18. **previous_cancellations**: Number of previous bookings by the guest that were canceled.

19. **previous_bookings_not_canceled**: Number of previous bookings by the guest that were not canceled.

20. **reserved_room_type**: The room type initially reserved by the guest.

21. **assigned_room_type**: The room type assigned to the guest (may differ from the reserved room type).

22. **booking_changes**: Number of changes made to the booking after the initial reservation.

23. **deposit_type**: Type of deposit made for the booking (e.g., No Deposit, Non Refund).

24. **agent**: ID of the travel agent who made the booking (if applicable).

25. **company**: ID of the company that made the booking (if applicable).

26. **days_in_waiting_list**: Number of days the booking was on a waiting list before confirmation.

27. **customer_type**: Type of customer (e.g., Transient, Group).

28. **adr**: Average Daily Rate – the average revenue per available room.

29. **required_car_parking_spaces**: Number of car parking spaces required by the guest.

30. **total_of_special_requests**: Total number of special requests made by the guest (e.g., extra bed).

31. **reservation_status**: Final status of the reservation (e.g., Check-Out, Canceled, No-Show).

32. **reservation_status_date**: Date when the reservation status was last updated.

```python
# creating numerical and categorical dataframes

cat_cols = [col for col in df.columns if df[col].dtype == 'O']
cat_cols
```

```python
cat_df = df[cat_cols]
cat_df.head()
```

```python
cat_df['reservation_status_date'] = pd.to_datetime(cat_df['reservation_status_date'])

cat_df['year'] = cat_df['reservation_status_date'].dt.year
cat_df['month'] = cat_df['reservation_status_date'].dt.month
cat_df['day'] = cat_df['reservation_status_date'].dt.day
```

```python
cat_df.drop(['reservation_status_date','arrival_date_month'] , axis = 1, inplace = True)
```

```python
cat_df.head()
```

```python
# printing unique values of each column
for col in cat_df.columns:
    print(f"{col}: \n{cat_df[col].unique()}\n")
```

```python
# encoding categorical variables

cat_df['hotel'] = cat_df['hotel'].map({'Resort Hotel' : 0, 'City Hotel' : 1})

cat_df['meal'] = cat_df['meal'].map({'BB' : 0, 'FB': 1, 'HB': 2, 'SC': 3, 'Undefined': 4})

cat_df['market_segment'] = cat_df['market_segment'].map({'Direct': 0, 'Corporate': 1, 'Online TA': 2, 'Offline TA/TO': 3,
                                                           'Complementary': 4, 'Groups': 5, 'Undefined': 6, 'Aviation': 7})

cat_df['distribution_channel'] = cat_df['distribution_channel'].map({'Direct': 0, 'Corporate': 1, 'TA/TO': 2, 'Undefined': 3,
                                                                       'GDS': 4})

cat_df['reserved_room_type'] = cat_df['reserved_room_type'].map({'C': 0, 'A': 1, 'D': 2, 'E': 3, 'G': 4, 'F': 5, 'H': 6,
                                                                   'L': 7, 'B': 8})

cat_df['deposit_type'] = cat_df['deposit_type'].map({'No Deposit': 0, 'Refundable': 1, 'Non Refund': 3})

cat_df['customer_type'] = cat_df['customer_type'].map({'Transient': 0, 'Contract': 1, 'Transient-Party': 2, 'Group': 3})

cat_df['year'] = cat_df['year'].map({2015: 0, 2014: 1, 2016: 2, 2017: 3})
```

```python
cat_df.head()
```

```python
num_df = df.drop(columns = cat_cols, axis = 1)
num_df.drop('is_canceled', axis = 1, inplace = True)
num_df
```

```python
num_df.var() # variance
```

```python
# normalizing numerical variables

num_df['lead_time'] = np.log(num_df['lead_time'] + 1)
num_df['arrival_date_week_number'] = np.log(num_df['arrival_date_week_number'] + 1)
num_df['arrival_date_day_of_month'] = np.log(num_df['arrival_date_day_of_month'] + 1)
num_df['agent'] = np.log(num_df['agent'] + 1)
num_df['company'] = np.log(num_df['company'] + 1)
num_df['adr'] = np.log(num_df['adr'] + 1)
```

```python
num_df.var()
```

```python
num_df['adr'] = num_df['adr'].fillna(value = num_df['adr'].mean())
```

```python
num_df.head()
```

```python
X = pd.concat([cat_df, num_df], axis = 1)
y = df['is_canceled']
```

```python
X.shape, y.shape
```

```python
# splitting data into training set and test set

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)
```

```python
X_test.head()
```
